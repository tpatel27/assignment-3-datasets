### Exploration & Visualization

The most important data in this dataset is the development related data like the type of developer, tools used, most used languages, databases,IDE's etc., all of this contains multiple data separated by a semicolon *;*, making it a bit harder to analyze directly

```python
survey_data.DevType.value_counts()
```

We can define a function which helps to us to reduce this lists into a single data-frame with single column for each.

```python
def split_multi_data(col_series):
    result_df = col_series.to_frame()
    options = []
    # Iterate over the column
    for idx, value  in col_series[col_series.notnull()].iteritems():
        # Break each value into list of options
        for option in value.split(';'):
            # Add the option as a column to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False
            # Mark the value in the option column as True
            result_df.at[idx, option] = True
    return result_df[options]
```

Now we can assign True or False value if the respondent has selected falls under the field.

```python
dev_data_df = split_multi_data(survey_data.DevType)
dev_data_df
```

Now we can summarize the above data in column wise total values.

```python
dev_totals = dev_data_df.sum().sort_values(ascending=False)
dev_totals
```

And now we can finally visualize the above data to see what category of DevType is the most preferred.

```python
plt.figure(figsize=(10,10))
plt.barh(dev_totals.index, dev_totals)
plt.title('DevType')
plt.gca().invert_yaxis()
```

Same way we can visualize other columns as well

```python
# We can combine the above formula and perform total of all the DF's in one go.
platform_totals = split_multi_data(survey_data.PlatformWorkedWith).sum().sort_values(ascending=False)
ide_totals = split_multi_data(survey_data.IDE).sum().sort_values(ascending=False)
```

### Combining two datasets together

We can combine two data together group them and create a visualization out of it, for example here we will compare the salaries of different regions in terms of USD.

```python
# Here we're using groupby() operation which involves combination of splitting the object, applying a function, and combining the results.
country_vs_salary = survey_data.groupby('Country')[['ConvertedSalary']].mean().sort_values('ConvertedSalary', ascending=False)
```

```python
# loc property is used to access a group of rows and columns by label(s)
country_vs_salary_df = country_vs_salary.loc[survey_data.Country.value_counts() > 250].head(15)
```

### 1. What programming languages are used most?
